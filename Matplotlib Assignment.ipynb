```python
#Question 2
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.xlabel("X - Axis")
plt.ylabel("y - Axis")
plt.title("Scatter Plot - Question 2")
plt.scatter(x,y)
plt.show()
```

```python
#Question 4
'''Bar plot is a kind of visualization technique and it is used when we are plotting numerical data against categorical data'''
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
```

```python
plt.barh(company, profit)
```
